# 5.1. Total Sales by Country 

```python
df2 = df.groupby('COUNTRY')['SALES'].sum().reset_index()
df2 = df2.sort_values(by='SALES',ascending=False)

plt.style.use('_mpl-gallery')
plt.figure(figsize=(15,7))
plt.bar(df2["COUNTRY"], df2["SALES"],color= "blue",width=0.3)
plt.title("Total Sales Across Different Countries", size=20)
plt.xlabel('Country', size= 20)
plt.ylabel('Sales', size = 20)
plt.xticks(rotation = 45)
current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:,.2f}M'.format(x / 1000000) for x in current_values])
plt.tight_layout
plt.show() 

```

# 5.2 Total Quantity Ordered by Country

```python
df3 = df.groupby('COUNTRY')['QUANTITYORDERED'].sum().reset_index()
df3 = df3.sort_values(by='QUANTITYORDERED', ascending=False)
plt.style.use('_mpl-gallery') 

plt.figure(figsize = (15,7))
plt.bar(df3['COUNTRY'], df3['QUANTITYORDERED'], color = 'violet',width = 0.3)
plt.title("Total Quantity ordered across different countries", size = 20)
plt.xlabel("Country", size = 20)
plt.ylabel("Quantity ordered", size = 20)
plt.xticks(rotation = 45)
current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:,.2f}K'.format(x / 1000) for x in current_values])
plt.tight_layout
plt.show() 
```

# 5.4 Total Sales Variation

```python
df_2 = df.groupby('YEAR_ID')['SALES'].sum().reset_index()
df_2['YEAR_ID'] = df_2['YEAR_ID'].astype(int)
plt.style.use('_mpl-gallery') 

plt.figure(figsize = (15,7))
plt.plot(df_2['YEAR_ID'], df_2['SALES'], color = 'red',linestyle ='--', linewidth = 5, marker = 'o', markersize = 10)

plt.title("Sales Variation", size = 20)
plt.xlabel("Year", size = 20)
plt.ylabel("Sales", size = 20)
plt.xticks(ticks=df_2['YEAR_ID'], labels=df_2['YEAR_ID'].astype(str))
current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:,.2f}M'.format(x / 1000000) for x in current_values])
plt.show() 
```

# 5.5 Top 10 Cities by Sales

```python
df_3 = df.groupby('CITY')['SALES'].sum().reset_index()
df_3 = df_3.sort_values(by='SALES', ascending=False)
df_4 = df_3.head(10)

plt.figure(figsize=(15, 7))
colors = plt.cm.Blues(df_4['SALES'] / df_4['SALES'].max())
bars = plt.bar(df_4['CITY'], df_4['SALES'], color=colors)

plt.title("Top 10 Cities By Sales", size=20)
plt.xlabel("City", size=15)
plt.ylabel("Sales", size=15)
current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:,.2f}M'.format(x / 1000000) for x in current_values])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 5.6 Dealsize Distribution

```python
sizes = df['DEALSIZE'].value_counts()
colors = ['blue', 'green', 'orange']
plt.figure(figsize=(3, 3))
plt.pie(sizes, labels=sizes.index, autopct='%1.1f%%',colors=sns.color_palette('Set2'), startangle = 140)
plt.axis('equal')
plt.title('Distribution of Small, Medium, and Large Dealsizes')
plt.show()
```

# 5.7 Sales by Country and Dealsize

```python
grouped_data = df.groupby(['COUNTRY', 'DEALSIZE'])['SALES'].sum().unstack()
colors = ['lightblue', 'lightgreen', 'orange'] 
grouped_data = grouped_data.loc[grouped_data.sum(axis=1).sort_values(ascending=False).index]
grouped_data.plot(kind='bar', stacked=True, color=colors, figsize=(10, 6))

plt.title('Sales by Country and Dealsize')
plt.xticks(rotation = 45)
plt.xlabel('Country')
plt.ylabel('Total Sales')
plt.show()
```

# 5.8 Top 10 Customers of all time

```python
df_3 = df.groupby('CUSTOMERNAME')['SALES'].sum().reset_index()
df_3 = df_3.sort_values(by='SALES', ascending=False)
df_4 = df_3.head(10)

plt.figure(figsize=(15, 7))
colors = plt.cm.Blues(df_4['SALES'] / df_4['SALES'].max())
bars = plt.bar(df_4['CUSTOMERNAME'], df_4['SALES'], color=colors)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x(), yval , yval)


plt.title("Top 10 Customers of all Time", size=20)
plt.xlabel("Customer Name", size=15)
plt.ylabel("Sales", size=15)
current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:,.2f}M'.format(x / 1000000) for x in current_values])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
